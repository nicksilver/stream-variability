```python
import numpy as np
import matplotlib.pyplot as plt
# from ggplot import *
import pandas as pd
```

```python
i = np.arange(50.0, 1201)
p = 50.0
w = 2.6
e0 = 400.0
```

```python
e_ip = 1.0 - ((1.0 + ((p+i)/e0)**w)**((1.0/w) - 1.0))*((p+i)/e0)**(w-1.0)
e_e0 = 1 - ((1.0 + (e0/(p+i))**w)**((1.0/w) - 1.0))*(e0/(p+i))**(w-1.0)

# e_i = (1.0 + (e0/(i + p))**w - (1.0 + (e0/(i + p))**w)**w**(-1.0))/(1.0 + (e0/(i + p))**w)
# e_e0 = 1.0 - (e0/(i + p))**(w - 1.0)*((e0/(i + p))**w + 1.0)**(1.0/w - 1.0)
# e_p = 1 - ((e0/(i + p))**w + 1)**(1/w - 1)  # same as e_i
```

```python
plt.plot(i, e_ip, i, e_e0) #, i, e_p)
plt.xlabel('Total Water')
plt.ylabel('Sensitivity')
plt.legend(['de/dip', 'de/de0'])  #, 'de/dp'])
plt.show()
```

## Compare distributions of change in PET

```python
np.random.seed(12)
e0_before = 50*np.random.randn(100)+80
e0_after = 50*np.random.randn(100)+150
p = 50*np.random.randn(100)+300
e_before = p*(1-(1 + (e0_before/p) - ((1 + (e0_before/p)**w)**(1/w))))
e_after = p*(1-(1 + (e0_after/p) - ((1 + (e0_after/p)**w)**(1/w))))
data = pd.DataFrame({'before':e_before, 'after':e_after})
dmelt = pd.melt(data[['before', 'after']])

ggplot(dmelt, aes(x='value', color='variable')) +\
    geom_density()
```

## Compare distributions fo change in P

```python
p_before = 50*np.random.randn(100)+800
p_after = 50*np.random.randn(100)+1000
e0 = 50*np.random.randn(100)+300
e_before = p_before*(1-(1 + (e0/p_before) - ((1 + (e0/p_before)**w)**(1/w))))
e_after = p_after*(1-(1 + (e0/p_after) - ((1 + (e0/p_after)**w)**(1/w))))
data = pd.DataFrame({'before':e_before, 'after':e_after})
dmelt = pd.melt(data[['before', 'after']])

ggplot(dmelt, aes(x='value', color='variable')) +\
    geom_density()
```
